```python
from bookworm import *
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12,9)

import pandas as pd
import numpy as np
import csv
import networkx as nx
```

```python
interaction_df = pd.read_csv('data/ASOIAF001_Character_Interaction_From_Excel_List.csv')
```

```python
Characters = ['Arya ', 'Balon ', 'Brienne ', 'Bran ', 'Catelyn ', 'Cersei ', 'Dany ', 'Eddard ', 'Jaime ', 'Joffrey ', 'Jon ', 'Lysa ', 'Ned ', 'Robb ', 'Robert ', 'Sansa ', 'Stannis ', 'Theon ', 'Tyrion ', 'Tywin ', 'Walder ', 'Daenerys ', 'Rhaegar ', 'Renly ', 'Rickon ', 'Varys ', 'Margaery ', 'Littlefinger ', 'Barristan ', 'Davos ', 'Samwell ']
```

```python
Houses = ['Baratheon ','Stark ','Lannister ','Tyrell ','Arryn ','Bolton ','Snow ','Tully ','Targaryen ','Martell ','Greyjoy ','Bolton ','Frey ','Mormont ']
```

```python
G = nx.from_pandas_edgelist(interaction_df,
                             source='source',
                             target='target', edge_attr=True)
```

```python
len(list(nx.enumerate_all_cliques(G)))
```

```python
#Weighted Pagerank
Pagerank = pd.Series(nx.pagerank(G, weight = "value")).sort_values(ascending=False)
Pagerank_df = Pagerank.to_frame()
Pagerank_df['rank'] = Pagerank_df[0].rank(ascending = False)
Pagerank_df.head()
Chars_PR = Pagerank_df.loc[ Characters ,: ]
Houses_PR = Pagerank_df.loc[ Houses ,: ]
print (Chars_PR)
print (Houses_PR)
```

```python
#Hits
a, b = nx.hits(G)
Hits = pd.Series(a).sort_values(ascending=False)
Hits_df = Hits.to_frame()
Hits_df['rank'] = Hits_df[0].rank(ascending = False)
Chars_Hits = Hits_df.loc[ Characters ,: ]
Houses_Hits = Hits_df.loc[ Houses ,: ]
print (Chars_Hits)
print (Houses_Hits)
```

```python
#Weighted Degree
WeightedDegree = pd.Series(sorted(G.degree(weight='value'), key=lambda x: x[1], reverse=True))
print(WeightedDegree[:30])
```

```python
#Degree
Degree = pd.Series(nx.degree_centrality(G)).sort_values(ascending=False)
Degree_df = Degree.to_frame()
Degree_df['rank'] = Degree_df[0].rank(ascending = False)
Chars_Degree = Degree_df.loc[ Characters ,: ]
Houses_Degree = Degree_df.loc[ Houses ,: ]
print (Chars_Degree)
print (Houses_Degree)
```

```python
#Closeness
Closeness = pd.Series(nx.closeness_centrality(G)).sort_values(ascending=False)
Closeness_df = Closeness.to_frame()
Closeness_df['rank'] = Closeness_df[0].rank(ascending = False)
Chars_Close = Closeness_df.loc[ Characters ,: ]
Houses_Close = Closeness_df.loc[ Houses ,: ]
print (Chars_Close)
print (Houses_Close)
```

```python
#Weighted Betweenness
Betweenness = pd.Series(nx.betweenness_centrality(G, weight = "value", normalized = True,endpoints = False)).sort_values(ascending=False)
Betweenness_df = Betweenness.to_frame()
Betweenness_df['rank'] = Betweenness_df[0].rank(ascending = False)
Chars_Btwn = Betweenness_df.loc[ Characters ,: ]
Houses_Btwn = Betweenness_df.loc[ Houses ,: ]
print (Chars_Btwn)
print (Houses_Btwn)
```

```python
#Weighted Eigenvector
WEigenvectors = pd.Series(nx.eigenvector_centrality(G, weight = "value")).sort_values(ascending=False) 
WEigenvectors_df = WEigenvectors.to_frame()
WEigenvectors_df['rank'] = WEigenvectors_df[0].rank(ascending = False)
Chars_WEV = WEigenvectors_df.loc[ Characters ,: ]
Houses_WEV = WEigenvectors_df.loc[ Houses ,: ]
print (Chars_WEV)
print (Houses_WEV)
```

```python
#Eigenvector
Eigenvectors = pd.Series(nx.eigenvector_centrality(G)).sort_values(ascending=False) 
Eigenvectors_df = Eigenvectors.to_frame()
Eigenvectors_df['rank'] = Eigenvectors_df[0].rank(ascending = False)
Chars_EV = Eigenvectors_df.loc[ Characters ,: ]
Houses_EV = Eigenvectors_df.loc[ Houses ,: ]
print (Chars_EV)
print (Houses_EV)
```
